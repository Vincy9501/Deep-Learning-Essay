```python
from sklearn import tree
from sklearn.datasets import load_wine # 生成数据集的类
from sklearn.model_selection import train_test_split
```

# 3. 划分训练集和测试集

```python
# 划分30%的数据作为测试集
Xtrain, Xtest, Ytrain, Ytest = train_test_split(wine.data, wine.target, test_size=0.3)
```

```python
Xtrain.shape # 训练集: 124行,13个标签
```

```python
Xtest.shape # 测试集: 54行,13个标签
```

# 4. 建模三部曲（实例化、训练模型、获取信息）

```python
clf = tree.DecisionTreeClassifier(criterion="entropy") # 建模
clf = clf.fit(Xtrain, Ytrain) # 训练模型
score = clf.score(Xtest, Ytest) # 返回预测的准确度
```

```python
score
```

# 5. 画树

```python
feature_name = ['酒精','苹果酸','灰','灰的碱性','镁','总酚','类黄酮','非黄烷类酚类','花青素','颜色强度','色调','od280/od315稀释葡萄酒','脯氨酸']

import graphviz
# 1. 输入已经训练好的模型
# 2. 特征名
# 3. 标签名
# 4. 填充颜色表示不同的分类
# 5. 圆角显示
dot_data = tree.export_graphviz(clf
                               ,feature_names = feature_name
                               ,class_names = ['琴酒', '雪莉', '贝尔摩德']
                                ,filled = True
                                ,rounded = True
                               )
graph = graphviz.Source(dot_data)

graph
```

# 决策树是随机的，如果要让模型稳定下来，需要random_state & splitter

```python
clf = tree.DecisionTreeClassifier(criterion="entropy",random_state=30)
clf = clf.fit(Xtrain, Ytrain)
score = clf.score(Xtest, Ytest) # 返回预测的准确度

score
```

```python
clf = tree.DecisionTreeClassifier(criterion="entropy"
                                 ,random_state=30
                                 ,splitter="random"
                                 )
clf = clf.fit(Xtrain, Ytrain) # 训练模型
score = clf.score(Xtest, Ytest)  # 返回预测的准确度

score
```

```python
import graphviz
dot_data = tree.export_graphviz(clf
                               ,feature_names= feature_name
                               ,class_names=["琴酒","雪莉","贝尔摩德"]
                               ,filled=True
                               ,rounded=True
                               )  
graph = graphviz.Source(dot_data)

graph
```

```python
# 我们的树对训练集的拟合程度如何？
score_train = clf.score(Xtrain, Ytrain)
score_train
```

# 如何确定最优的剪枝参数？

```python
import matplotlib.pyplot as plt
test = []
for i in range(10):
    clf = tree.DecisionTreeClassifier(max_depth=i+1
                                     ,criterion="entropy"
                                     ,random_state=30
                                     ,splitter="random"
                                     )
    clf = clf.fit(Xtrain, Ytrain)
    score = clf.score(Xtest, Ytest)
    test.append(score)
plt.plot(range(1,11),test,color="red",label="max_depth")
plt.legend()
plt.show()
```

# 4.2 交叉验证

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor

iris = load_iris()
regressor = DecisionTreeRegressor(random_state=0) # 回归
# cv 把数据分成10份，每次取一份做测试集，剩下是训练集，循环十次
# score r2 或者 MSE
cross_val_score(regressor, iris.data, iris.target, cv=10, 
                scoring = "neg_mean_squared_error")
```
